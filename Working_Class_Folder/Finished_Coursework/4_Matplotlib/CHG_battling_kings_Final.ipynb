```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import csv
```

```python
# Read CSV
battle = os.path.join("Resources/got.csv")

battle_df = pd.read_csv(battle)
battle_df

```

```python
# Get attacker and defender data

atk = battle_df['attacker_king'].value_counts()

df = battle_df['defender_king'].value_counts()
```

```python
atk
```

```python
df
```

```python
# Get total battle data
kings = atk.add(df,fill_value=0)
```

```python
# Configure plot and ticks
kings.plot(kind='bar',facecolor='red')
```

```python
# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.xlabel("King")
plt.ylabel("Number of Battles Participated In")
```

```python
# Show plot
plt.show()
```
