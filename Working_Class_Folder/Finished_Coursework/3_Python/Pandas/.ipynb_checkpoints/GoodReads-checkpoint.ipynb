```python
# Import Dependencies
import os
import csv
import pandas as pd
```

```python
# Make a reference to the books.csv file path
books = os.path.join("PandaResources/books.csv")
# Import the books.csv file as a DataFrame
goodreads_df = pd.read_csv(books, encoding="ISO-8859-1")

goodreads_df.head()
```

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame
# Only keep: "isbn", "original_publication_year", "original_title", "authors",
# "ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"

cleanreads_df = goodreads_df[["isbn","original_publication_year", "original_title", "authors","ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"]]

cleanreads_df
```

```python
# Rename the headers to be more explanatory
readablereads_df = cleanreads_df.rename(columns={"isbn":"ISBN",
                                                "original_publication_year":"Publication Year",
                                               "original_title":"Original Title",
                                               "authors":"Authors",
                                               "ratings_1":"One Star Reviews",
                                               "ratings_2":"Two Star Reviews",
                                               "ratings_3": "Three Star Reviews",
                                               "ratings_4": "Four Star Reviews",
                                               "ratings_5": "Five Star Reviews"})
```

```python
# Push the remade DataFrame to a new CSV file
readablereads_df.to_csv("GoodReads1_Output.csv", index=False, header=True)
```

```python
books_updated = os.path.join ("GoodReads1_Output.csv")

goodreads2_df = pd.read_csv(books_updated,encoding="ISO-8859-1")
```

```python
goodreads2_df.describe()
```

```python
unique_count = goodreads2_df["Authors"].nunique()
unique_count
```

```python
earliest_year = goodreads2_df["Publication Year"].min()
earliest_year
```

```python
latest_year = goodreads2_df["Publication Year"].max()
latest_year
```

```python
star1 = goodreads2_df["One Star Reviews"].sum()
star2 = goodreads2_df["Two Star Reviews"].sum()
star3 = goodreads2_df["Three Star Reviews"].sum()
star4 = goodreads2_df["Four Star Reviews"].sum()
star5 = goodreads2_df["Five Star Reviews"].sum()

total_review = star1 + star2 + star3 + star4 + star5

total_review
```

```python
summary = {"Summary":{
    "Unique Authors":unique_count,
    "Earliest Publication":earliest_year,
    "Latest Publication":latest_year,
    "Total Reviews Across All Authors":total_review}
}

summary_df = pd.DataFrame(summary)
```

```python
summary_df
```

```python
summary_df.to_csv("Summary_of_GoodReads.csv",index=True,header=True)
```
