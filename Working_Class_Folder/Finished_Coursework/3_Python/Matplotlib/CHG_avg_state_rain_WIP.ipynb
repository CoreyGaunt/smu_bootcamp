```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import csv
```

```python
# Load in csv

rain = os.path.join("avg_rain_state.csv")

rain_df = pd.read_csv(rain)



raindf = rain_df.rename(columns={"State":"State",
                       "Inches":"Inches",
                       "Millimetres":"Millimeters",
                       "Rank":"Rank"})

raindf.head()
```

```python
# creating x-axis labels
states = [x for x in rain_df['State']]
```

```python
# creating y-axis labels
inches = [x for x in rain_df['Inches']]
```

```python
plt.plot(states,inches,marker="o",color="blue",alpha=0.5)
plt.xlabel("State")
plt.ylabel("Rainfall in Inches")
plt.title("Average Rainfall (In) Across States")
plt.axis("equal")
plt.xticks(rotation="vertical")

plt.show()
```
