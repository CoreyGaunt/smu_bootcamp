```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
dd = np.array([9,8,90,85,80,70,70,65,55,60,70,65,50])
rg = np.array([75,70,60,65,60,45,55,50,40,40,35,35,30])
```

```python
duration = np.arange(0,130,10)
```

```python
plt.plot(duration,dd)
plt.plot(duration,rg)
plt.hlines(0,0,10,alpha=.25)
dangerdrop, = plt.plot(duration,dd,marker='v',color='red',label="Danger Drop")
railgun, = plt.plot(duration,rg,marker='^',color='green',label="RailGun")
plt.xlim(0, 120)
plt.ylim(0, 100)
plt.legend(loc='upper left')
plt.xlabel("Coaster Runtime(s)")
plt.ylabel("Speed (MPH)")
plt.title("Coaster Speed Over Time")
plt.grid()
plt.show()
```
