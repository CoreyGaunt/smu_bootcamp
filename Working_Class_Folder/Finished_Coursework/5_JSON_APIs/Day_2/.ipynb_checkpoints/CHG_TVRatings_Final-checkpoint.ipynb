```python
#Dependencies
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#list of tv show titles to query
tv_shows = ["Altered Carbon", "Grey's Anatomy", "This is Us", "The Flash", "Vikings", "Shameless", "Arrow", "Peaky Blinders", "Dirk Gently"]

ratings = []

# make iterative requests to TVmaze search endpoint

url = 'http://api.tvmaze.com/singlesearch/shows?q='

for show in tv_shows:
    response = requests.get(url + show).json()
    ratings.append(response['rating']['average'])
    
ratings
    
    
```

```python
# create dataframe

ratingsDF = {'Title':tv_shows,
             'TV Rating':ratings}

TV_rat = pd.DataFrame(ratingsDF)

tvRatings_sorted = TV_rat.sort_values(by='TV Rating',ascending=False)
tvRatings_sorted
```

```python
# use pandas to create a bar chart from the dataframe

tvRatings_sorted.plot('Title', 'TV Rating', kind='bar', figsize=(10,5), rot=45)
plt.show()
plt.tight_layout()
```
