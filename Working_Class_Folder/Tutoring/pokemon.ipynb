```python
# Dependencies
import pandas as pd
import os
import csv
```

```python
# Save file path to variable
poke_file = os.path.join("Pokemon.csv")

poke_df = pd.read_csv(poke_file)
```

```python
# Read with Pandas
poke_df.head()
```

```python
# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"

sub_df = poke_df[["Type 1","HP","Attack","Sp. Atk","Sp. Def","Speed"]]

sub_df.head()
```

```python
# Create a dataframe of the average stats for each type of pokemon.
stat_df = sub_df.groupby("Type 1").mean().round(2)

stat_df

```

```python
# Calculate the total power level of each type of pokemon by summing all of the stats together.
stat_df["Power"] = stat_df["HP"] + stat_df["Attack"]+ stat_df["Sp. Atk"]+ stat_df["Sp. Def"]+ stat_df["Speed"]
stat_df

# Place the results into a new column.
# Hint: Research the sum method from pandas.
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.

sorted_df = stat_df.sort_values("Power",ascending=False)
sorted_df
```
